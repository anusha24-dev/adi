```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
text="""HAPPY NEW YEAR MY FAMILY
PARVATHAMMA SHIVANNA,NANJUNDAPPA SHANTHAMMA ANUSHA
MANJULA NAGESH MOULYA SANDEEP❤️❤️❤️❤️"""

wordcloud=WordCloud(width=800,height=400,background_color='blue').generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.title("happynew year")
plt.show()

           
                    
      
```
